```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```
f = pd.read_csv('data2.csv')
f.head()
```

```
#check nilai null pada data
f.isnull().sum()
```

```
f.info()
```

```
categorical = ['Jenis_Tinggal','Alat_Transportasi','Penerima_KPS','Pekerjaan_Ayah','Penghasilan_Ayah','Pekerjaan_Ibu','Penghasilan_Ibu','Rombel_Saat_Ini','Penerima_KIP','Layak_PIP']
for i in categorical:
  plt.figure(figsize=(20,6))
  ax = sns.countplot(x=i,data=f);
  ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",fontsize=6);
```

```
kategori =  ['Jenis_Tinggal','Alat_Transportasi','Penerima_KPS','Pekerjaan_Ayah','Penghasilan_Ayah','Pekerjaan_Ibu','Penghasilan_Ibu','Rombel_Saat_Ini','Penerima_KIP','Layak_PIP']
for j in kategori:
  f[j] = f[j].fillna(f[j].mode().values[0])
f.head()
```

```
#check nilai null setelah handle missing value
f.isnull().sum()
```

```


f['Penerima_KPS'] = f['Penerima_KPS'].replace({'Ya': 1, 'Tidak': 2})
f['Penerima_KIP'] = f['Penerima_KIP'].replace({'Ya': 1, 'Tidak': 2})
f['Layak_PIP'] = f['Layak_PIP'].replace({'Ya': 1, 'Tidak': 2})
f['Jenis_Tinggal'] = f['Jenis_Tinggal'].replace({'Bersama orang tua':1, 'Wali':2, 'Asrama':3, 'Kost':4, 'Lainnya':5})
f['Alat_Transportasi'] = f['Alat_Transportasi'].replace({'Sepeda motor':1, 'Ojek':2, 'Jalan kaki':3, 'Kendaraan pribadi':4, 'Mobil Pribadi':5, 'Kereta Api':6, 'Mobil/bus antar jemput':7,
                                 'Angkutan umum/bus/pete-pete':8})
f['Pekerjaan_Ayah'] = f['Pekerjaan_Ayah'].replace({'PNS/TNI/Polri':1, 'Wiraswasta':2, 'Sudah Meninggal':3, 'Karyawan Swasta':4, 'Wirausaha':5, 'Pedagang Kecil':6, 'Petani':7,
                                 'Buruh':8, 'Pensiunan':9, 'Tidak Bekerja':10,'Peternak':11, 'Pedagan Besar':12, 'Lainnya':13})
f['Penghasilan_Ayah'] = f['Penghasilan_Ayah'].replace({'Rp. 2,000,000 - Rp. 4,999,999':1, 'Tidak Berpenghasilan':2, 'Rp. 5,000,000 - Rp. 20,000,000':3, 'Rp. 500,000 - Rp. 999,999':4, 'Rp 1.000.000 - Rp 2.000.000':5, 'Rp. 1,000,000 - Rp. 1,999,999':6, 'Lebih dari Rp 2.000.000':7,'Kurang dari Rp 1.000.000':8, 'Lebih dari Rp. 20,000,000':9, 'Kurang dari Rp. 500,000':10})
f['Pekerjaan_Ibu'] = f['Pekerjaan_Ibu'].replace({'PNS/TNI/Polri':1, 'Pedagang Kecil':2, 'Tidak Bekerja':3, 'Wirausaha':4, 'Karyawan Swasta':5, 'Lainnya':6, 'Petani':7,
                                 'Wiraswasta':8, 'Peternak':9, 'Buruh':10,'Sudah Meninggal':11, 'Pensiunan':12})
f['Penghasilan_Ibu'] = f['Penghasilan_Ibu'].replace({'Rp. 2,000,000 - Rp. 4,999,999':1, 'Tidak Berpenghasilan':2, 'Rp. 5,000,000 - Rp. 20,000,000':3, 'Rp. 500,000 - Rp. 999,999':4, 'Rp 1.000.000 - Rp 2.000.000':5, 'Rp. 1,000,000 - Rp. 1,999,999':6, 'Lebih dari Rp 2.000.000':7,'Kurang dari Rp 1.000.000':8, 'Lebih dari Rp. 20,000,000':9, 'Kurang dari Rp. 500,000':10})
kategori = ['Pekerjaan_Ayah','Pekerjaan_Ibu','Rombel_Saat_Ini']
for i in kategori:
  num = LabelEncoder()
  f[i] = num.fit_transform(f[i].astype('str'))
f
```

```
#create correlation with heatmap
corr = f.corr(method = 'pearson')

#convert correlation to numpy array
mask = np.array(corr)

#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(12,12)
sns.heatmap(corr, mask = mask, vmax = 0.9, square = True, annot = True)
```
